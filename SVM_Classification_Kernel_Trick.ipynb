# SVM Classification kernel trick

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# Create dataset
from sklearn.datasets import make_circles
X, y = make_circles(100, factor=0.1, noise=0.1)
plt.scatter(X[:,0], X[:,1], c = y, s = 50, cmap='bwr')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
classifier = SVC(kernel='linear')
classifier.fit(X_train, y_train.ravel())
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
plot_decision_regions(X_train, y_train, clf = classifier)
```

```python
def plot_3d_plot(X,y):
    r = np.exp(-(X**2).sum(1))
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:,0], X[:,1], r, c=y, s= 100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(X,y)
```

```python
# using kerel as rbf
rbf_classifier = SVC(kernel='rbf')
rbf_classifier.fit(X_train, y_train)
y_pred = rbf_classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
plot_decision_regions(X,y,rbf_classifier)
```

```python
# using kerel as poly
ploy_classifier = SVC(kernel='poly', degree=2)
ploy_classifier.fit(X_train, y_train)
y_pred = ploy_classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
plot_decision_regions(X,y,ploy_classifier)
```

```python
X_new = np.exp(-(X**2))
plt.scatter(X_new[:,0], X_new[:,1], c = y, cmap='bwr')
```

```python
# using kerel as poly
ploy_classifier = SVC(kernel='poly', degree=2)
ploy_classifier.fit(X_new, y)
y_pred = ploy_classifier.predict(X_new)
```

```python
plot_decision_regions(X_new,y,ploy_classifier)
```
